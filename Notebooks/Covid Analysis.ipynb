
### Covid - Análise
Vamos focar em cinco indicadores-chave para a análise dos dados de COVID-19. Abaixo, contém os indicadores-chave que vamos olhar.

Indicadores-Chave Selecionados
- Total de Casos Confirmados por Região
- Total de Mortes por Região
- Taxa de Mortalidade por Região
- Crescimento Diário de Novos Casos
- Crescimento Diário de Novas Mortes

```
%sql
USE covid;
```

```
%sql
SELECT * FROM covid.covid_data LIMIT 5;
```

#### Análise dos Indicadores

```
%sql
SELECT
  country_name,
  SUM(total_confirmed_cases) AS total_cases
FROM
  covid.covid_data
GROUP BY
  country_name
ORDER BY
  total_cases DESC;
```

Insight Esperado: Identificar as regiões mais afetadas em termos de número absoluto de casos confirmados. Isso pode ajudar a determinar as áreas com maior disseminação do vírus.

```
%sql
SELECT
  country_name,
  SUM(total_confirmed_deaths) AS total_deaths
FROM
  covid.covid_data
GROUP BY
  country_name
ORDER BY 
  total_deaths DESC;
```

Insight Esperado: Avaliar o impacto letal do COVID-19 por região, o que é fundamental para entender a gravidade da pandemia em diferentes locais.

```
%sql
SELECT 
    country_name, 
    ROUND (SUM(total_confirmed_deaths) / SUM(total_confirmed_cases) * 100, 2) AS mortality_rate
FROM 
    covid.covid_data
GROUP BY 
    country_name
order BY mortality_rate DESC;
```

Insight Esperado: Determinar a taxa de mortalidade por região, fornecendo um indicador da letalidade do vírus nas diferentes áreas geográficas.

```
%sql
SELECT 
    country_name, 
    date, 
    total_confirmed_cases - LAG(total_confirmed_cases, 1) OVER (PARTITION BY country_name ORDER BY date) AS daily_new_cases
FROM 
    covid.covid_data;
```

Insight Esperado: Analisar como o número de novos casos está mudando ao longo do tempo, o que pode indicar se a pandemia está se acelerando ou desacelerando em diferentes regiões.

```
%sql
SELECT 
    country_name, 
    date, 
    total_confirmed_deaths - LAG(total_confirmed_deaths, 1) OVER (PARTITION BY country_name ORDER BY date) AS daily_new_deaths
FROM 
    covid.covid_data;
```
